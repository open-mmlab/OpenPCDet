```python
from numba import jit

import random

@jit(nopython=True)
def monte_carlo_pi(nsamples):
    acc = 0
    for i in range(nsamples):
        x = random.random()
        y = random.random()
        if (x ** 2 + y ** 2) < 1.0:
            acc += 1
    return 4.0 * acc / nsamples
```

```python
import torch
import torch.nn.functional as F

# inp = torch.randn(1, 1, 20, 20)
inp = torch.arange(5*5).view(1, 1, 5, 5).float()
out_h = 5
out_w = 5

new_h = torch.linspace(-1, 1, out_h).view(-1, 1).repeat(1, out_w)
new_w = torch.linspace(-1, 1, out_w).repeat(out_h, 1)
grid = torch.cat((new_h.unsqueeze(2), new_w.unsqueeze(2)), dim=2)
print(grid)
grid = grid.unsqueeze(0)
print(grid.shape)
output = F.grid_sample(inp, grid=grid, mode='bilinear')
print(output - inp)
```

```python
N, H, W = 4, 32, 64

grid_x = torch.arange(W, dtype=torch.float)
grid_y = torch.arange(H, dtype=torch.float)
grid_x, grid_y = torch.meshgrid(grid_y, grid_x)
grid_x = grid_x / (W -1) * 2 - 1
grid_y = grid_y / (H - 1) * 2 -1

# grid coor shape is N 1 H W 2
grid_coor = torch.stack((grid_x, grid_y), dim=-1).view(1, H, W, 2).repeat(4, 1, 1, 1)
print(grid_coor.shape)

input_feature = torch.ones(N, 4, H, W)
output = F.grid_sample(input_feature, grid_coor, mode='bilinear')
print(output.shape)


print(output - input_feature)
```
